```python
import umap
import umap.plot
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
dreams = pd.read_csv('C:\\Users\\bookw\\Dropbox\\Capstone\\DreamBERT\\00_data\\dream_text_corpus')
dream = dreams['dreams']
```

```python
person = dreams['titles']
```

```python
reducer = umap.UMAP()
```

```python
#sentence embedding generated by huggingface transformer
#t0pp for zeroshot
#Umng has clustering code
vectorizer = CountVectorizer(min_df=5, stop_words='english')
word_doc_matrix = vectorizer.fit_transform(dream)
```

```python
# cosine: all together on one side with no discernable clusters
embedding = umap.UMAP(
    # number of dimensions to return (2 for now)
    n_components = 2, 
    # number of neighbors it uses (local v global)
    n_neighbors = 5, 
    # minimum distance between neighbors
    min_dist = 0,
    # controls how distance is computed
    metric = 'cosine').fit(word_doc_matrix)

f = umap.plot.points(embedding)
```

```python
# braycurtis: a blob with someone off in the corner
embedding = umap.UMAP(
    # number of dimensions to return (2 for now)
    n_components = 2, 
    # number of neighbors it uses (local v global)
    n_neighbors = 5, 
    # minimum distance between neighbors
    min_dist = 0,
    # controls how distance is computed
    metric = 'braycurtis').fit(word_doc_matrix)

f = umap.plot.points(embedding)
```
